```python
from sklearn import datasets
bunch = datasets.load_breast_cancer()
```

```python
data = bunch.data
labels = bunch.target  #良性惡行的結果
```

```python
print(data)
print(type(data))
print(data.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, labels,
                                                    test_size=0.3,
                                                    shuffle=True,
                                                    stratify=labels)
```

```python
#Original
from sklearn.ensemble import RandomForestClassifier
randomForest = RandomForestClassifier(n_estimators=100, class_weight='balanced', n_jobs=-1, verbose=1) #分別為: 100棵樹 自動調整權重 cpu全部用上 訓練過程的資料列出來
```

```python
# 進行訓練
randomForest.fit(X_train, Y_train)
```

```python
# 進行預測
acc = randomForest.score(X_test, Y_test)

print('Accuracy:',acc)
```

## 特徵選擇-過濾法
#### 對各個特徵進行評分(依據特徵與target的相關性、特徵的發散程度-標準差)，設定選擇的特徵個數或門檻

### 過濾方式:
* 移除變異數低的特徵
* 單變量特徵選擇

```python
# 過濾方式1
#移除變異數低的特徵
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold=0.01) #只要特徵的變異數小於0.01，他就會被刪除
selector = selector.fit(X_train, Y_train)  #根據訓練資料和結果建立selection
X_train = selector.transform(X_train)  #透過transform刪除或留下特徵
X_test = selector.transform(X_test)
#建立模型
from sklearn.ensemble import RandomForestClassifier
randomForest = RandomForestClassifier(n_estimators=100, class_weight='balanced', n_jobs=-1, verbose=1)
```

```python
randomForest.fit(X_train, Y_train)
```

```python
# 進行預測
acc = randomForest.score(X_test, Y_test)
print('Accuracy:',acc)

#準確度提升
```

```python
# 過濾方式2
#單變量特徵選擇
from sklearn.feature_selection import SelectKBest, chi2  #SelectKBest: 只留下依據評分指標評分後前k名的特徵
selector = SelectKBest(chi2,k=10)  #留下前10個特徵比較好的(依據chi2評分標準)去建構selector
selector = selector.fit(X_train, Y_train)
X_train = selector.transform(X_train)
X_test = selector.transform(X_test)
#建立模型
from sklearn.ensemble import RandomForestClassifier
randomForest = RandomForestClassifier(n_estimators=120, criterion='entropy', class_weight='balanced', n_jobs=-1, verbose=1)
```

```python
randomForest.fit(X_train, Y_train)
```

```python
# 進行預測
acc = randomForest.score(X_test, Y_test)
print('Accuracy:',acc)

#準確度提升
```
