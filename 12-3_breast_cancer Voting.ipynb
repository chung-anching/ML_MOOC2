## 以威斯康辛乳癌數據集，實做投票法(Voting)

```python
# 載入資料
from sklearn import datasets
bunch = datasets.load_breast_cancer()
```

```python
data = bunch.data
labels = bunch.target
```

```python
# 切割資料
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(data, labels, test_size=0.3,shuffle=True,stratify=labels)
```

```python
# 建立模型
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

clf1 = LogisticRegression(C=1e5)
clf2 = RandomForestClassifier(n_estimators=50, class_weight='balanced')
clf3 = SVC(C=1e5,kernel='linear', gamma='scale', class_weight='balanced')
```

```python
from sklearn.ensemble import VotingClassifier
votingClassifier = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3)], voting='hard') # estimators是list，也是一連串的預測器，以名稱和預測器模型配對格式列出
# voting='hard'指定投票的方法為"硬投票": 每個模型輸出他覺得最可能的類別，再多數決(軟投票是對所有類別產生機率分布的向量)
```

```python
# 進行訓練
votingClassifier.fit(X_train, Y_train)
```

```python
# 進行預測
acc = votingClassifier.score(X_test, Y_test)

print('Accuracy:',acc)
```

```python
predictions = votingClassifier.predict(X_test)
```
