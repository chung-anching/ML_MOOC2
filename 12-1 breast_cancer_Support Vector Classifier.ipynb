```python
from sklearn import datasets
bunch = datasets.load_breast_cancer()
```

```python
#資料取得與切割
data = bunch.data
labels = bunch.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, labels, test_size=0.3, shuffle=True, stratify=labels)
```

```python
from sklearn.svm import SVC
```

```python
#Original
svc = SVC() #建立模型
```

```python
# 進行訓練
svc.fit(X_train, Y_train)
```

```python
# 進行預測
acc = svc.score(X_test, Y_test)
print('Accuracy:',acc)
```

```python
#調整模型參數
svc = SVC(C=1e5, kernel='linear', gamma='scale', class_weight='balanced') #class_weight='balanced':因為資料有病跟沒病的數量不平均，'balanced'做平衡
```

```python
# 進行訓練
svc.fit(X_train, Y_train)
# 進行預測
acc = svc.score(X_test, Y_test)
print('Accuracy:',acc)
```

```python
# 用RFE,遞迴特徵選擇
svc = SVC(C=1e5, kernel='linear', gamma='scale', class_weight='balanced')
from sklearn.feature_selection import RFE
selector = RFE(estimator=svc, n_features_to_select=27) #從30個特徵中挑選較重要的27個
selector = selector.fit(X_train, Y_train)
X_train = selector.transform(X_train)
X_test = selector.transform(X_test)
```

```python
# 進行訓練
svc.fit(X_train, Y_train)
```

```python
# 進行預測
acc = svc.score(X_test, Y_test)
print('Accuracy:',acc)  #精準很多
```

```python
import pickle   #將模型存成pickle檔案(使用 Pickle 模組保存資料)

# 儲存Model
with open('svc.pickle', 'wb') as f:   # 'svc.pickle':檔名；'wb':可以以二元的形式寫進去(writable binary)
    pickle.dump(svc, f)   #呼叫dump真的把它寫出去

# 讀取Model
with open('svc.pickle', 'rb') as f:  #read binary讀進來
    svc = pickle.load(f)
```

```python
# 測試讀入的Model
prediction = svc.predict(X_test[0:1])
print(prediction)
```

## 使用 pickle 儲存、讀取資料 範例

```python
import pickle

my_dict = {
    'a': 1,
    'b': 2,
    'c': 3,
}

with open('test.pickle', 'wb') as f:
    pickle.dump(my_dict, f)
    
# 我們可以使用 "dump()" 函式來儲存我們的資料，以上方程式碼為例，便是我的 "my_dict" 的資料。
```

```python
# Load
with open('test.pickle', 'rb') as f:
    new_dict = pickle.load(f)

print(new_dict)

# 再來紀錄使用 "load()" 讀取儲存好的 pickle 資料。
```
