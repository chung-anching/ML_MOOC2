```python
# 載入資料
from sklearn import datasets
bunch = datasets.load_breast_cancer()
```

```python
data = bunch.data
labels = bunch.target
```

```python
# 切割資料
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(data, labels, test_size=0.3,shuffle=True,stratify=labels)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC
baggingClassifier = BaggingClassifier(base_estimator=SVC(), n_estimators=10)  # base_estimator:基礎預測器，是sklearn的分類器或回歸器(預設是決策樹)
baggingClassifier.fit(X_train, Y_train)                      # n_estimators:基礎預測器的數量(預設10個)，預測器越多 整體模型變異越小
```

```python
predictions = baggingClassifier.predict(X_test)
accuracy = baggingClassifier.score(X_test, Y_test)
```

```python
print(predictions)
print(accuracy)
```
